```python
def isReflexive(relation):
    ## The function utilizes an integer variable countTrue 
    ## to count each time the equation is reflexive 
    countTrue=0
    ## The two for loops iterate through the dictionary by its keys (x) 
    ## and each item (or value) in the list within each of the keys (y)
    for x in relation:
        for y in range(len(relation[x])):
            ## This boolean checks to see if the keys contains an item equal
            ## to itself (which would mean the relation is reflexive)
            ## If it does, countTrue increments 
            if(x==relation[x][y]): 
                countTrue+=1
    ## The final boolean compares the number of keys that contained an item
    ## equal to themselves 
    ## If the two numbers are equal, it indicates every key in the relation
    ## is reflexive
    if countTrue==len(relation):
        return True
    else: return False
    
def isSymmetric(relation):
    ## The first for loop iterates all key 
    for x in relation:
        ## Variable truth checks if each relation is symmetric for all values
        ## It resets as false at the beginning of every loop 
        truth=False
        ## The second for loop iterates through all definitions (values) in the
        ## current key being iterated through
        ## This resets truth as False for the beginning of each comparison
        for y in relation[x]:
            truth=False
            ## This boolean checks if the value of the current key is contained
            ## in the list of the key with the same value as y  
            if (x in relation[y]):
                ## If x is in the list, then the relation is symmetric in that
                ## instance
                truth=True
            ## If at any point the relation is not symmetric, then it cannot be
            ## fully symmetric, and thus the function returns False
            if (truth==False):
                return False
    ## At the end of iterating through all keys, the code checks the symmetry a
    ## final time and as before if any part is not symmetric, returns false
    if(truth==False):
        return False
    ## However, if the code finds truth to still be True, the function returns 
    ## True (that the inputted relation was symmetric)
    else:
        return True
    
def isAntiSymmetric(relation):
    ## The first for loop iterates through the entire dictionary 
    ## with the keys being x and the key items' lists(y)
    for x,y in relation.items():
        ## The next for loop iterates through each item of y's list
        for i in range(len(y)):
            ## The boolean checks if the dictionary (relation) contains
            ## the value in y's list and ensures the item is not equal to x
            if y[i] in relation and y[i]!=x:
                ## This for loop iterates through the keys and key item lists 
                for n,m in relation.items():
                    ## This boolean finds the key with value equal to the
                    ## current item in x's list being iterated through
                    if n==y[i]:
                        ## The for loops checks to see if the value of x is
                        ## contained within the list of key y
                        ## It checks if yRx
                        for j in range(len(m)):
                            if m[j]==x:
                                ## If x is in the key y's list, the relation
                                ## cannot be asymmetric and thus the function
                                ## returns False
                                return False
    ## If the function iterates through every key's list and finds no instance
    ## of value x being contained within the list of key y, then there is no
    ## instance in the relation where xRy AND yRx and thus, it is asymmetric
    ## So the function returns True
    return True

def isTransitive(relation):
    ## The three for loops iterate first through the keys of the dictionary
    ## (x), then the items within each key (y), and finally the items within 
    ## the keys of the dictionary equal to value y (z)
    for x in relation:
        for y in relation[x]:
            for z in relation[y]:
                ## This boolean checks if value y is contained in relation[x]
                ## Because if y is contained, xRy is True
                if(y in relation[x]):
                    ## Next the function checks if the value z is contained 
                    ## within the key equal to value y b/c if so, yRz
                    if(z in relation[y]):
                        ## The final boolean states that if value z is not 
                        ## contained within key x's list, the relation is 
                        ## not transitive and the function returns False
                        ## Essentially, it checks if xRz
                        ## If there is no xRz, the relation is not transitive
                        if(z not in relation[x]):
                            ## Thus if the relation is not transitive, the
                            ## function returns False
                            return False
    ## If the function finds that the relation never had xRy and yRz but not
    ## xRz, then it returns True as the relation must then be transitive
    return True
```
