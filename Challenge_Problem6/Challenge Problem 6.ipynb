```python
def isFunction(domain, codomain, relation):
    ## This code section ensures only one y output for every one x input
    ## 'count' keeps track of how many times the domain element is seen in the x element of the relation list
    ## The two for loops check every element of the domain against every x element in the 'relation' list
    ## If the two elements are equivalent, then the count increases by 1
    ## If 'count' is equal to more than 1, there are more 2 or more outputs for a single x input which means it's not a function
    ## The final boolean checks to see if the value in the domain was never used at all
    ## If it wasn't, the function is not well defined and the program returns False
    for i in range(len(domain)): 
        count = 0
        for j in range(len(relation)): 
            if (domain[i] == relation[j][0]): 
                count+=1                
            if (count >1):
                return False
        if (count<1):
            return False
            
    ## This code ensures that no x's are out of the domain
    ## The two for loops check every x element in the relation list against every element in the domain
    ## 'count' keeps track of how many times an element of the domain is equal to an x element in the relation list
    ## If 'count'= 0 then there is an x element in the relation list not found in the domain, meaning it's outside the domain
    ## This means it's not a function
    for i in range(len(relation)):
        count=0
        for j in range(len(domain)):
            
            if (domain[j]==relation[i][0]):
                count+=1
           
        if (count==0):
            return False
    
    ## This code ensures that no y's are out of the codomain
    ## The two for loops check every y element in the relation list against every element in the codomain
    ## 'count' keeps track of how many times an element of the domain is equal to a y element in the relation list
    ## If 'count'= 0 then there is a y element in the relation list not found in the codomain, meaning it's outside the codomain
    ## This means it's not a function
    for i in range(len(relation)):
        count=0
        for j in range(len(codomain)):
            
            if (codomain[j]==relation[i][1]):
                count+=1
           
        if (count==0):
            return False
    return True    



def isOneToOne(domain, codomain, relation):
    ## This code tests the input to ensure it is a function (for understanding on its logic see isFunction())
    if(isFunction(domain, codomain, relation)!=True):
        return False
    
    ## This code tests if a function is one to one
    ## The for loops check every y element of the relation list against one another 
    ## The little i!=j part just makes sure the code doesn't compare the same element against itself
    ## If the y elements of two different relations are equal, than the function must not be one to one
    for i in range(len(relation)): 
        for j in range(len(relation)):
            if ((i!=j)and(relation[i][1]==relation[j][1])):
                return False
    else: return True    
        
        
        
def isOnto(domain, codomain, relation):
    ## This code tests the input to ensure it is a function (for understanding on its logic see isFunction())
    if(isFunction(domain, codomain, relation)!=True):
        return False
    
    ## This code tests if a function is onto
    ## The for loops check every element of the codomain against every y element of the relation list
    ## If the y element and element of the codomain are equal, 'count' is incremented by 1
    ## If 'count' is equal to 0 after checking every y element of the relation list against the element of the codomain...
    ## Then there must not be an x input such that the output is y, therefore it is not onto
    for i in range(len(codomain)):
        count=0
        for j in range(len(relation)):
            if (codomain[i]==relation[j][1]):
                count+=1
        if (count==0):
            return False
    return True



def inverse(domain, codomain, relation):
    invertedList=[]
    ## This code tests the input to ensure it is a function (for understanding on its logic see isFunction())
    ## It also checks if it is onto and one to one (explanations are included in isOnto() and isOneToOne())
    ## If it is a function, onto, and one to one, the code will fill the invertedList with the x and y values...
    ## From the original relation list, but it will swap their places (thus inverting the function)
    if((isFunction(domain, codomain, relation))and(isOnto(domain, codomain, relation))and(isOneToOne(domain, codomain, relation))):
        for i in range(len(relation)):
            invertedList+=[[relation[i][1],relation[i][0]]]
        return(invertedList)
            
    ## This code tests if it's not a function by checking to see if isFunction() is false
    ## If it is not a function, it raises the appropriate exception
    if(isFunction(domain, codomain, relation)==False):
        raise Exception('Input is not a function')
        
    ## This code tests if it's a function but not Onto and One to One by seeing if isFunction() is true...
    ## And isOnto() or isOneToOne() is false
    ## If it is not both onto and one to one, the code raises the appropriate exception
    if(isFunction(domain,codomain,relation)and((isOnto(domain, codomain, relation)==False)or(isOneToOne(domain, codomain, relation)==False))):
        raise Exception('Input is not invertible')
```
